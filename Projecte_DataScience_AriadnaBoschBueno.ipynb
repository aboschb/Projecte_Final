```python
import tweepy
import csv

# Indiquem les credencials de Twitter per autenticar-nos (compte Basic)
consumer_key = 'XXXXXXXXXXXXXXXXXXXXXXX' # (Hem ocultat les credencials)
consumer_secret = 'XXXXXXXXXXXXXXXXXXXXXXXXX' 
access_token = 'XXXXXXXXXXXXXXXXXXXXXXX' 
access_token_secret = 'XXXXXXXXXXXXXXXXXXXXXXXXXXXX' 

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)

# Cerca de tweets que mencionen "@rodalies", retornant fins a 100 resultats en format complet
tweets = api.search_tweets(q="@rodalies", count=50, tweet_mode='extended')

# Obrim un fitxer CSV per escriure
with open('tweets.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Escriu les capçaleres del CSV
    writer.writerow(['Nom_usuari', 'Data', 'Text', 'Num_Likes', 'Reposts'])

    # Processem i escrivim la informació dels tweets al fitxer CSV
    for tweet in tweets:
        nom_usuari = tweet.user.screen_name
        data = tweet.created_at
        text = tweet.full_text.replace('\n', ' ')  # Evitem salts de línia dins del camp de text
        num_likes = tweet.favorite_count
        reposts = tweet.retweet_count

        # Escrivim les dades del tweet al fitxer CSV
        writer.writerow([nom_usuari, data, text, num_likes, reposts])
```

### 3. Preprocessament i Transformació de les Dades

Com hem explicat abans, el dataframe conté el nom dels usuaris i volem mantenir la seva privacitat, així que per començar a mostrar les dades farem una còpia del mateix en la qual es mostren aquestes dades de forma oculta.

Per garantir la privacitat dels usuaris quan compartim dades públicament, implementarem un mètode per substituir els noms d'usuari originals, que apareixen al nostre conjunt de dades com noms de compte de X/Twitter, amb identificadors únics creats per nosaltres. Aquesta mesura cautelar implica la creació d'un id_usuari per a cada usuari únic: @rodalies es manté inalterat per a reconèixer el compte oficial, mentre que els altres noms d'usuari són reemplaçats per identificadors anònims com user1, user2, user3, etc. Això permet analitzar i compartir les dades sense comprometre la identitat dels usuaris, assegurant que puguem mantenir l'ètica de la privacitat mentre treballem amb informació potencialment sensible.

```python
# Inicialitzem un diccionari per al seguiment dels ids d'usuari
user_ids = {}

# Creem una funció per obtenir l'id d'usuari basant-nos en el nom d'usuari
def get_user_id(nom_usuari):
    # Si el nom d'usuari és @rodalies, assignem directament 'rodalies'
    if nom_usuari == '@rodalies':
        return 'rodalies'
    # Si el nom d'usuari no és @rodalies, comprovem si ja té un id assignat
    if nom_usuari not in user_ids:
        # Assignem un nou id d'usuari 
        user_ids[nom_usuari] = f'user{len(user_ids) + 1}'
    # Retornem l'id d'usuari existent o el nou assignat
    return user_ids[nom_usuari]

# Apliquem la funció a cada fila per crear la nova columna 'id_usuari'
df['id_usuari'] = df['nom_usuari'].apply(get_user_id)

# Per ocultar els noms d'usuari originals, els substituim amb 'xxx'
# Això no canvia les dades, només com es mostren en el notebook
df['nom_usuari'] = 'xxx'

# Mostrem una mostra del dataframe
df.sample(5)
```

A continuació transformarem les dades de la columna imatges, de floats a booleanes, això ens permetrà fer possibles anàlisis de l'engagement dels tweets depenent de si contenen imatges o no.

```python
# Transformem la columna 'imatges' de floats a booleanes
df['imatges'] = df['imatges'].apply(lambda x: False if x == 0.0 else True)

df['imatges'].dtype
```

Com podem veure, hem transformat correctament les dades de la columna 'imatges' de float a booleanes, de manera que els valors que es mostraran seran True o False. Ara farem una altra transformació: passarem les variables id_tweet, id_main_tweet i id_usuari a categòriques, per evitar confusions en el nostre anàlisi.

```python
df['id_tweet'] = df['id_tweet'].astype('object')
df['id_main_tweet'] = df['id_main_tweet'].astype('object')
df['id_usuari'] = df['id_usuari'].astype('object')

print(df['id_tweet'].dtype)
print(df['id_main_tweet'].dtype)
print(df['id_usuari'].dtype)
```

Ara analitzarem si la longitud del tweet influeix en l'engagement que rep. Per visualitzar els resultats farem servir diagrames de violí. Per tal de facilitar la visualització, les dades de la variable 'tweet_long' s'han categoritzat en 3 tipus: tweet curt, mig i llarg. Originalment, Twitter tenia una limitació de 140 caràcters, que després es va ampliar a 280. La categoria de menys de 50 caràcters pot representar tweets que són molt directes o que utilitzen principalment hashtags o enllaços. La categoria intermèdia reflecteix l'ús més comú de l'espai de tweets abans de l'expansió del límit de caràcters, i els tweets de més de 100 caràcters aprofiten plenament la capacitat ampliada.

```python
# Definim els límits de les categories de longitud de text
# els textos de menys de 50 caràcters seran 'curt'
# entre 50 i 100 caràcters seran 'mig',
# i 'llarg' els de més de 100 caràcters.
bins = [0, 50, 100, float('inf')]
labels = ['curt', 'mig', 'llarg']
df['text_category'] = pd.cut(df['text_long'], bins=bins, labels=labels)

# Ara creem els violinplots per a cada variable d'engagement
# Configurem l'estil general dels gràfics
sns.set(style="whitegrid")

# Creem un violinplot per a num_likes
plt.figure(figsize=(10, 6))
sns.violinplot(x='text_category', y='num_likes', data=df, color='coral')
plt.title('Distribució de "m\'agrada" segons la llargària del text')
plt.xlabel('Categoria del text')
plt.ylabel('Nombre de "m\'agrada"')
plt.savefig('violinplot1.png')
plt.show()

# Creem un violinplot per a num_comments
plt.figure(figsize=(10, 6))
sns.violinplot(x='text_category', y='num_comments', data=df, color='coral')
plt.title('Distribució de comentaris segons la llargària del text')
plt.xlabel('Categoria del text')
plt.ylabel('Nombre de comentaris')
plt.savefig('violinplot2.png')
plt.show()

# Creem un violinplot per a num_visualitzacions
plt.figure(figsize=(10, 6))
sns.violinplot(x='text_category', y='num_visualitzacions', data=df, color='coral')
plt.title('Distribució de visualitzacions segons la llargària del text')
plt.xlabel('Categoria del text')
plt.ylabel('Nombre de visualitzacions')
plt.savefig('violinplot3.png')
plt.show()
```

Els gràfics de violí mostren la distribució de "m'agrada", comentaris i visualitzacions en funció de la llargària del text dels tweets, categoritzats com a curts, mitjans i llargs. 

Per entendre els gràfics, l'amplada del "cos" de cada violí indica la densitat de tweets que reben un cert nombre d'engagement, amb la part més ampla representant una major concentració de dades. Els pics fora del cos principal del violí (línies o antenes) representen valors extrems en l'engagement.

Això pot indicar que la seva brevetat no aporta prou engagement com perquè la resta d'usuaris responguin positivament. Pel que fa al nombre de comentaris, els tweets de llargària mitjana tornen a ser els que més comentaris generen. Per últim, pel que fa al nombre de visualitzacions, 
novament els tweets mitjans són els que més visualitzacions aconsegueixen. Ara eliminarem les URLs,mencions i símbols dels hashtags en tot el dataset:

```python
import re

# Definim una funció per eliminar mencions, URLs i el símbol # dels hashtags
def remove_mentions_urls_and_hash(text):
    if isinstance(text, str) and text.strip():
        # Eliminem URLs
        text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
        # Eliminem mencions
        text = re.sub(r'@\w+', '', text)
        # Eliminem només el símbol # dels hashtags
        text = re.sub(r'#', '', text)
        return text.strip()
    else:
        return ""

# Apliquem la funció al dataframe
df['text_netejat'] = df['text'].apply(remove_mentions_urls_and_hash)

# Desactivem la limitació de caràcters en la sortida de Pandas
pd.set_option('display.max_colwidth', None)

# Sobrescrivim el fitxer Excel original amb el dataframe modificat
df.to_excel('df_processat.xlsx', index=False)
```

A continuació volem crear una nova columna anomenada 'idioma', que identifica la llengua del text de cada tweet, ja que tenim tweets en català, castellà i anglès. Prèviament, en la consola hem instal.lat la llibreria necessària per fer-ho:

```python
from langdetect import detect, LangDetectException

def detect_language(text):
    # Comprovar si el text és una cadena no buida
    if isinstance(text, str) and text.strip():
        try:
            # Detectem l'idioma del text
            lang = detect(text)
            if lang == 'ca':
                return 'CAT'
            elif lang == 'es':
                return 'ESP'
            elif lang == 'en':
                return 'ENG'
            else:
                return 'ALTRES'  # Per a altres idiomes
        except LangDetectException:
            return 'INDETERMINAT'  # En cas que no es pugui determinar l'idioma
    else:
        # Per valors nuls o no text
        return 'INDETERMINAT'

# Apliquem la funció al dataframe per crear la nova columna 'idioma'
df['idioma'] = df['text'].apply(detect_language)
```

```python
from textblob import TextBlob

def detect_language(text):
    # Comprovar si el text és una cadena no buida
    if isinstance(text, str) and text.strip():
        try:
            # Utilitzar TextBlob per a la detecció d'idioma
            textblob_language = TextBlob(text).detect_language()

            # Mapejar els codis de TextBlob als teus codis d'idioma desitjats
            language_mapping = {
                'ca': 'CAT',
                'es': 'ESP',
                'en': 'ENG'
            }

            # Obtenir l'etiqueta corresponent o mantenir el valor original si no coincideix cap
            detected_language = language_mapping.get(textblob_language, textblob_language)

            return detected_language
        except Exception:
            return text  # Mantenir el valor original en cas d'error
    else:
        # Mantenir el valor original (inclosos els buits) per a valors nuls o sense text
        return text

# Aplica la funció de detecció d'idioma al dataframe només als valors 'ALTRES'
df['idioma'] = df[df['idioma'] == 'ALTRES']['idioma'].apply(detect_language)
```

A continuació utilitzarem el model en anglès per al text en anglès (depenent del valor de la columna 'idioma') i el model en espanyol per al text tant en català com en castellà. Després comprovarem els resultats.

```python
# Creem una funció que selecciona el model basat en l'idioma i elimina noms propis
def remove_named_entities_based_on_language(text, language):
    if language == 'ENG':
        doc = nlp_eng(text)
    elif language in ['CAT', 'ESP']:  # Indiquem que pel text en català i castellà s'utilitzi el model en espanyol
        doc = nlp_spa(text)
    else:
        return text  # Retorna el text sense modificar si l'idioma no és reconegut
    
    # Filtrem les entitats que no són noms propis (PERSON)
    filtered_tokens = [token.text for token in doc if not (token.ent_type_ == "PERSON")]
    
    # Unim els tokens filtrats en una cadena
    return ' '.join(filtered_tokens)

# Apliquem la funció al DataFrame
df['text_sense_noms'] = df.apply(lambda row: remove_named_entities_based_on_language(row['text_netejat'], row['idioma']), axis=1)
```

```python
# Inicialitzem una llista per acumular els noms de persona extrets
extracted_person_names = []

# Definim la funció per eliminar noms de persona en ESP o CAT i acumular els noms extrets
def remove_person_names_for_esp_cat(text, language):
    # Comprova si l'idioma és ESP o CAT
    if language in ['ESP', 'CAT']:
        # Processa el text utilitzant el model en espanyol
        doc = nlp_spa(text)
        
        # Construeix el text netejat excluint els noms de persona i acumula els noms extrets
        filtered_text = []
        for ent in doc.ents:
            if ent.label_ == "PER":
                extracted_person_names.append(ent.text)
        return ''.join([token.text_with_ws if not any(ent.start <= token.i < ent.end for ent in doc.ents if ent.label_ == "PER") else '' for token in doc])
    else:
        # Retorna el text sense modificar si l'idioma no és ESP o CAT
        return text

# Apliquem la funció de manera condicional a la columna 'text_netejat'
df['text_sense_noms'] = df.apply(lambda row: remove_person_names_for_esp_cat(row['text_netejat'], row['idioma']), axis=1)

# Visualitzem la llista de noms de persona extrets
print("Noms de persona extrets:", set(extracted_person_names))
```

A continuació ja podem utilitzar el model per a fer el reconeixement d'entitats nomenades:

```python
from transformers import pipeline
import warnings

# Ignorem totes les advertències de categoria DeprecationWarning
warnings.filterwarnings('ignore', category=DeprecationWarning)

# Creem la pipeline NER amb el model específic per a català
ner_pipeline = pipeline("token-classification", model="projecte-aina/roberta-base-ca-cased-ner")

def processar_text(text):
    # Aplica la pipeline NER al text
    entitats = ner_pipeline(text)
    noms_extrets = []
    text_original = text
    for entitat in entitats:
        if entitat['entity'].startswith('B-PER') or entitat['entity'].startswith('I-PER'):
            nom = entitat['word'].replace('Ġ', '')
            # Eliminar el nom del text amb una expressió regular per a ser insensible a majúscules i minúscules
            text = re.sub(r'\b' + re.escape(nom) + r'\b', '', text, flags=re.IGNORECASE)
            noms_extrets.append(nom)
    return text.strip(), noms_extrets


# Apliquem la funció només als tweets en català i crear les noves columnes
df['resultat'] = df.apply(lambda fila: processar_text(fila['text_netejat']) if fila['idioma'] == 'CAT' else (fila['text_netejat'], []), axis=1)

# Continuem amb la separació dels resultats en noves columnes com abans
df[['text_sense_noms', 'noms_eliminats']] = pd.DataFrame(df['resultat'].tolist(), index=df.index)
df.drop('resultat', axis=1, inplace=True)
```

Ara revisarem un exemple de l'output de ner_pipeline per entendre millor l'estructura de les dades que retorna. Això ens pot ajudar a identificar com accedir correctament a la informació sobre les entitats nomenades:

```python
exemple_text = "Això és un exemple amb un nom com Jordi."
print(ner_pipeline(exemple_text))
```

El resultat ens mostra que el model ha identificat correctament "Jordi" com a entitat de tipus 'B-PER' (que significa "Beginning of a person's name" en el context de reconeixement d'entitats nominades, indicant l'inici d'un nom de persona), amb una alta confiança (score: 0.99827325).

L'output també mostra que la paraula reconeguda comença amb el caràcter Ġ, que és un artefacte de com el model tokenizer RoBERTa tracta els espais abans de les paraules. Això confirma que quan processem aquests resultats per eliminar noms de persona del text, necessitarem tenir en compte i eliminar aquest caràcter.

```python
def processar_text(text):
    # Apliquem la pipeline NER al text
    entitats = ner_pipeline(text)
    noms_extrets = []
    for entitat in entitats:
        if entitat['entity'] == 'B-PER':
            nom = entitat['word'].replace('Ġ', '')  # Netegem el prefix
            text = re.sub(r'\b' + re.escape(nom) + r'\b', '', text, flags=re.IGNORECASE)  # Elimina el nom del text
            noms_extrets.append(nom)
    return text.strip(), noms_extrets

# Apliquem la funció a la columna corresponent només per als tweets en català
df['resultat'] = df.apply(lambda fila: processar_text(fila['text_netejat']) if fila['idioma'] == 'CAT' else (fila['text_netejat'], []), axis=1)

# Separem el resultat en dues noves columnes: 'text_sense_noms' i 'noms_eliminats'
df[['text_sense_noms', 'noms_eliminats']] = pd.DataFrame(df['resultat'].tolist(), index=df.index)

# Eliminem la columna 'resultat' si ja no la necessites
df.drop('resultat', axis=1, inplace=True)
```

```python
def format_label(label):
    """Donem format a dues etiquetes específiques."""
    labels_to_change = ["Manca informació", "Queixa general"]
    return label.replace(" ", "\n") if label in labels_to_change else label

comptes_tematica.index = [
    format_label(label) for label in comptes_tematica.index
]

# Creem el gràfic circular
fig, ax = plt.subplots(figsize=(10, 8))

# Definim una funció que només retorna els percentatges si són >= 3%
def custom_autopct(pct):
    return ('{:.1f}%'.format(pct)) if pct >= 3 else ''

# Plotejem les dades. Passarem les etiquetes a la funció 'pie'.
patches, texts, autotexts = ax.pie(
    comptes_tematica,
    labels=comptes_tematica.index,
    autopct=custom_autopct,
    startangle=140
)

# Amaguem els textos dels percentatges i les etiquetes si són menors del 3%
for i, pct in enumerate(comptes_tematica / comptes_tematica.sum() * 100):
    if pct < 3:
        autotexts[i].set_visible(False)
        texts[i].set_visible(False)

# Afegim la llegenda amb ajustaments per evitar la superposició
ax.legend(patches, comptes_tematica.index, title="Temàtiques", loc="center left", bbox_to_anchor=(1.1, 0, 0.5, 1))

# Mostrem el gràfic
plt.show()
```

Per comparar resultats, ara farem una identificació de temàtiques mitjançant LDA. Provarem amb els tweets en castellà:

```python
from gensim import corpora, models
import re
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

# Filtrem el df per obtenir només els tweets en espanyol
df_esp = df[df['idioma'] == 'ESP']

# Preprocessem el text per a l'espanyol
stemmer = SnowballStemmer('spanish')
stop_words = set(stopwords.words('spanish'))

# Apliquem tokenització i neteja, aplicant stemming i eliminant stopwords
texts_esp = [
    [stemmer.stem(word) for word in re.findall(r'\b\w\w+\b', document.lower()) if word not in stop_words]
    for document in df_esp['text_sense_noms']
]

# Creem un diccionari de paraules i la matriu Document-Term per a l'espanyol
dictionary_esp = corpora.Dictionary(texts_esp)
corpus_esp = [dictionary_esp.doc2bow(text) for text in texts_esp]

# Entrenem el model LDA per a l'espanyol
lda_model_esp = models.LdaModel(corpus_esp, num_topics=10, id2word=dictionary_esp, passes=15)

# Mostrem els temes per a l'espanyol
topics_esp = lda_model_esp.print_topics(num_words=4)
for topic in topics_esp:
    print(topic)
```

### 5.3 Freqüència de paraules

En primer lloc eliminarem les stopwords dels nostres tweets. Pel català farem el següent:

```python
# Cree una funció per eliminar stopwords d'un text
def remove_stopwords(text):
    doc = nlp(text)
    filtered_sentence = ' '.join([token.text for token in doc if not token.is_stop])
    return filtered_sentence

# Apliquem la funció només als tweets en català
df.loc[df['idioma'] == 'CAT', 'text_sense_stopwords'] = df.loc[df['idioma'] == 'CAT', 'text_sense_noms'].apply(remove_stopwords)
```

```python
# Mostrem primers resultats de la columna 'text_sense_stopwords' per veure els canvis
print(df['text_sense_stopwords'].head())
```

Com podem veure, s'han eliminat els articles, pronoms, etc. del text. Ara farem una lematització:

```python
# Creem una funció per a la lemmatització del text
def lemmatize_text(text):
    doc = nlp(text)
    # Retorna una cadena de text amb els lemmes separats per espais
    lemmatized_text = ' '.join([token.lemma_ for token in doc])
    return lemmatized_text

# Apliquem la funció de lemmatització a la columna 'text_sense_stopwords'
df.loc[df['idioma'] == 'CAT', 'text_lemmatized'] = df.loc[df['idioma'] == 'CAT', 'text_sense_stopwords'].apply(lemmatize_text)

# Mostrem els resultats
print(df.loc[df['idioma'] == 'CAT', 'text_lemmatized'].head())
```

Observem que, en alguns casos, la lemmatització pot no ser completament òptima. Un exemple d'això és la transformació de 'gràcies' a 'gràcia'. Des d'un punt de vista sintàctic, aquesta conversió pot semblar correcta, ja que la lemmatització busca reduir les paraules a la seva forma base. No obstant això, aquesta transformació pot alterar el significat original de la frase. 'Gràcies' s'utilitza com a expressió d'agraïment, mentre que 'gràcia' té connotacions diferents, per exemple "fer gràcia". Aquest exemple subratlla un dels desafiaments del NLP: la dificultat de mantenir el significat precís i la intenció darrere de les paraules en el seu context quan s'apliquen processos automàtics com la lemmatització. Això demostra la importància de considerar el context i els matisos lingüístics quan es treballa amb eines de NLP.

Ara farem un càlcul de les freqüències de paraules en català:

```python
from collections import Counter

# Preprocessem el text per extreure lemmes en català
def preprocess_text(text):
    doc = nlp(text)
    lemmas = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct and not token.is_space]
    return lemmas

# Excloem els tweets publicats per l'usuari "Rodalies" i creem una còpia independent per evitar l'avís
df_usuaris = df[df['id_usuari'] != 'rodalies'].copy()

# Apliquem el preprocessament als tweets en català dels usuaris (excloent Rodalies)
df_cat_usuaris = df_usuaris[df_usuaris['idioma'] == 'CAT'].copy()  # Ens assegurem de treballar amb una còpia
df_cat_usuaris['text_lemmatized'] = df_cat_usuaris['text_netejat'].apply(preprocess_text)

# Concatenem totes les llistes de lemmes en una sola llista
all_lemmas_cat = sum(df_cat_usuaris['text_lemmatized'], [])

# Calculem la freqüència de cada lemma
lemma_freq_cat = Counter(all_lemmas_cat)

# Convertim el comptador a un DataFrame per a una millor visualització i anàlisi
df_lemma_freq_cat = pd.DataFrame(lemma_freq_cat.items(), columns=['Lemma', 'Frequency']).sort_values(by='Frequency', ascending=False).reset_index(drop=True)

# Mostrem les paraules més freqüents
print(df_lemma_freq_cat.head(10))
```

Veiem que "el" i "de" no estan incloses en la llista de stopwords, perquè apareixen aqui. Anem doncs a incloure-les a la llista:

```python
# Afegim paraules a la llista de stopwords, si no són reconegudes per defecte
for word in ['el', 'de']:
    nlp.Defaults.stop_words.add(word)
    nlp.vocab[word].is_stop = True
```

Ara actualitzarem la columna text_sense_stopwords per assegurar que les noves stopwords personalitzades siguin considerades. 

```python
def remove_custom_stopwords(text):
    doc = nlp(text)
    filtered_sentence = ' '.join([token.text for token in doc if not token.is_stop])
    return filtered_sentence

# Apliquem la funció per eliminar les noves stopwords del text original
df.loc[df['idioma'] == 'CAT', 'text_sense_stopwords'] = df.loc[df['idioma'] == 'CAT', 'text_sense_noms'].apply(remove_custom_stopwords)
```

Ara haurem d'actualitzar la columna text_lemmatized per reflectir els canvis realitzats a la columna text_sense_stopwords:

```python
# Carreguem el model de llenguatge català i actualitzem la llista de stopwords
nlp = spacy.load('ca_core_news_sm')
nlp.Defaults.stop_words.add('el')
nlp.vocab['el'].is_stop = True
nlp.Defaults.stop_words.add('de')
nlp.vocab['de'].is_stop = True

def preprocess_and_lemmatize(text):
    doc = nlp(text)
    # Filtrem tokens que no són stopwords i no són puntuació, i després lemmatitzem
    lemmas = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    # Filtrem explícitament paraules com 'el' i 'de' després de la lemmatització
    lemmas_filtered = [lemma for lemma in lemmas if lemma not in ['el', 'de']]
    return ' '.join(lemmas_filtered)  # Retorna com a cadena de text

# Apliquem la funció actualitzada a 'text_sense_stopwords' i guardem en 'text_lemmatized'
df.loc[df['idioma'] == 'CAT', 'text_lemmatized'] = df.loc[df['idioma'] == 'CAT', 'text_sense_stopwords'].apply(preprocess_and_lemmatize)
```

El que estem veient és que "el" i "de" segueixen incloent-se en la llista de paraules, malgrat les hem incloses com a stopwords en la nostra llista. Les exclourem específicament doncs:

```python
def preprocess_and_lemmatize(text):
    doc = nlp(text)
    # Retorna una llista de lemmes filtrant explícitament 'el' i 'de', a més de les stopwords i puntuació
    lemmas_filtered = [token.lemma_ for token in doc if token.lemma_ not in ['el', 'de'] and not token.is_stop and not token.is_punct]
    return lemmas_filtered 

# Apliquem la funció actualitzada i ens assegurem que es generen llistes de lemmes
df.loc[df['idioma'] == 'CAT', 'text_lemmatized'] = df.loc[df['idioma'] == 'CAT', 'text_sense_stopwords'].apply(preprocess_and_lemmatize)
```
